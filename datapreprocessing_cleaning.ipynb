Preprocessing of data

```python
import sklearn
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
import pandas as pd
import seaborn as sns

df = fetch_openml('titanic',version=1,as_frame=True)['data']
#print(df.info())
# print(df.isnull().sum())
sns.set()
missing_Value = pd.DataFrame((df.isnull().sum()/len(df))*100) # will give percentage of missing values
missing_Value.plot(kind = 'bar', title = 'missing values', ylabel = 'percentage')
df.drop(['body'], axis =1, inplace = True)
print(f'size of dataset is {df.shape}')
plt.show()
```

Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
fruits = pd.DataFrame({

	'weight (in KGs)':[1,3,4,2],
	'cost (in INR)':[120,200,150,100]},
	index=['apple','mango','orange','grapes'])

fruits1 = pd.DataFrame(scaler.fit_transform(fruits),columns=['weight (in KGs)','cost (in INR)'],
    index= ['apple','mango','orange','grapes'])
fruiscale = fruits.plot.scatter(x='cost (in INR)',y='weight (in KGs)',color=['red','blue','green','orange'],
    marker='*',s=80, label='Before scaling')
fruits1.plot.scatter(x='weight (in KGs)',y='cost (in INR)',color=['red','blue','green','orange'],
marker='o',s=60, label='After scaling',ax = fruiscale)
plt.axhline(0,color = 'red', alpha = 0.2)	
plt.axhline(0,color = 'red', alpha = 0.2)	
plt.show()
```
